```python
sayi=int(input("Bir sayı girin: "))
```

```python
cift_toplam=0
tek_toplam=0
```

```python
for i in range(1,sayi):
    if(i%2==0):
        cift_toplam+=i
    else:
        tek_toplam+=i
```

```python
print("Çift sayıların toplamı:{0}".format(cift_toplam))
print("Tek sayıların toplamı:{0}".format(tek_toplam))
```
